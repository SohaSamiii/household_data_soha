```python
def df_to_x_y3(df,window_size=5):
  df_as_np=df.to_numpy()
  x=[]
  y=[]
  for i in range(len(df_as_np)-window_size):
    row = [r for r in df_as_np [i:i+window_size]]
    x.append(row)
    label = df_as_np[i+window_size][0],df_as_np[i+window_size][1],df_as_np[i+window_size][2],df_as_np[i+window_size][3],df_as_np[i+window_size][4],df_as_np[i+window_size][5]
    y.append(label)
  return np.array(x),np.array(y)
```

```python
x3, y3 = df_to_x_y3(pv2_df)
x3.shape, y3.shape
```

```python
x3_train,y3_train=x3[:6000],y3[:6000]
x3_val,y3_val = x3[6000:7000],y3[6000:7000]
x3_test,y3_test = x3[7000:],y3[7000:]
```

```python
pv_training = np.mean(x3_train[:,:,0])
pv_std = np.std(x3_train[:,:,0])
pv_training1 = np.mean(x3_train[:,:,1])
pv_std1 = np.std(x3_train[:,:,1])
pv_training2 = np.mean(x3_train[:,:,2])
pv_std2 = np.std(x3_train[:,:,2])
pv_training3 = np.mean(x3_train[:,:,3])
pv_std3 = np.std(x3_train[:,:,3])
pv_training4 = np.mean(x3_train[:,:,4])
pv_std4 = np.std(x3_train[:,:,4])
pv_training5 = np.mean(x3_train[:,:,5])
pv_std5 = np.std(x3_train[:,:,5])


```

```python
def preprocess3(x):
  x[:,:,0] = (x[:,:,0]-pv_training)/pv_std
  x[:,:,1] = (x[:,:,1]-pv_training1)/pv_std1
  x[:,:,2] = (x[:,:,2]-pv_training2)/pv_std2
  x[:,:,3] = (x[:,:,3]-pv_training3)/pv_std3
  x[:,:,4] = (x[:,:,4]-pv_training4)/pv_std4
  x[:,:,5] = (x[:,:,5]-pv_training5)/pv_std5



  return x

```

```python
def preprocess_output3(y):
  y[:,0] = (y[:,0] - pv_training) / pv_std
  y[:,1] = (y[:,1] - pv_training1) / pv_std1
  y[:,2] = (y[:,2] - pv_training2) / pv_std2
  y[:,3] = (y[:,3] - pv_training3) / pv_std3
  y[:,4] = (y[:,4] - pv_training4) / pv_std4
  y[:,5] = (y[:,5] - pv_training5)/ pv_std5

  return y
```

```python
preprocess3(x3_train)
preprocess3(x3_val)
preprocess3(x3_test)
```

```python
preprocess_output3(y3_train)
preprocess_output3(y3_val)
preprocess_output3(y3_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.metrics import MeanAbsoluteError as mse
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Activation
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.metrics import MeanAbsoluteError
from tensorflow.keras.layers import Dense, LSTM, Activation
from tensorflow.keras.layers import GRU
from tensorflow.keras.optimizers import RMSprop

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense

model_cnn = Sequential()
model_cnn.add(InputLayer(input_shape=(5, 10)))  # ورودی با ابعاد (زمان، ویژگی‌ها)
model_cnn.add(Conv1D(filters=32, kernel_size=2, activation='relu', padding='same'))  # لایه کانولوشن با 32 فیلتر و اندازه هسته 2
model_cnn.add(MaxPooling1D(pool_size=2, padding='same'))  # لایه Pooling با اندازه Pool 2
model_cnn.add(Conv1D(filters=64, kernel_size=2, activation='relu', padding='same'))  # لایه کانولوشن با 64 فیلتر و اندازه هسته 2
model_cnn.add(MaxPooling1D(pool_size=2, padding='same'))  # لایه Pooling با اندازه Pool 2
model_cnn.add(Flatten())  # تخت کردن خروجی کانولوشن برای ورودی به لایه‌های Dense
model_cnn.add(Dense(16, activation='relu'))  # لایه Dense با 16 واحد و تابع فعال‌سازی ReLU
model_cnn.add(Dense(16, activation='relu'))  # لایه Dense با 16 واحد و تابع فعال‌سازی ReLU
model_cnn.add(Dense(16, activation='relu'))
model_cnn.add(Dense(16, activation='relu'))  # لایه Dense با 16 واحد و تابع فعال‌سازی ReLU
model_cnn.add(Dense(6, activation='linear'))  # لایه Dense با 8 واحد و تابع فعال‌سازی خطی

model_cnn.summary()  # نمایش ساختار مدل
```

```python
cp5 = ModelCheckpoint('model_cnn/model_checkpoint.keras', save_best_only=True)
```

```python
import tensorflow as tf
from tensorflow.keras import backend as K

def r2_score(y_true, y_pred):
    # میانگین مقادیر واقعی
    y_true_mean = K.mean(y_true)

    # مجموع مربعات خطا
    ss_res = K.sum(K.square(y_true - y_pred))

    # مجموع مربعات کل
    ss_tot = K.sum(K.square(y_true - y_true_mean))

    # R^2 score
    r2 = 1 - (ss_res / (ss_tot + K.epsilon()))
    return r2

model_cnn.compile(
    loss=MeanSquaredError(),
    optimizer=RMSprop(learning_rate=0.0001),
    metrics=[r2_score]
)
```

```python
model_cnn.fit(x3_train,y3_train,validation_data=(x3_val,y3_val), epochs=100,callbacks=[cp5])
```

```python
# پیش‌بینی‌ها
y_train_pred = model_cnn.predict(x3_train)
y_val_pred = model_cnn.predict(x3_val)

# پس‌پردازش پیش‌بینی‌ها

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# محاسبه MAE و RMSE
mae_train = mean_absolute_error(y3_train, y_train_pred)
rmse_train = np.sqrt(mean_squared_error(y3_train, y_train_pred))

mae_val = mean_absolute_error(y3_val, y_val_pred)
rmse_val = np.sqrt(mean_squared_error(y3_val, y_val_pred))

# نمایش نتایج
print("Training Metrics:")
print(f"Mean Absolute Error (Train): {mae_train}")
print(f"Root Mean Squared Error (Train): {rmse_train}")

print("\nValidation Metrics:")
print(f"Mean Absolute Error (Validation): {mae_val}")
print(f"Root Mean Squared Error (Validation): {rmse_val}")
```

```python
# پیش‌بینی با مدل GRU
gru_predictions_train = model_cnn.predict(x3_train)  # خروجی پیش‌بینی‌های آموزش
gru_predictions_val = model_cnn.predict(x3_val)     # خروجی پیش‌بینی‌های اعتبارسنجی
gru_predictions_test = model_cnn.predict(x3_test)   # خروجی پیش‌بینی‌های تست
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# تعداد ویژگی‌ها
n_features = y3_train.shape[1]

# لیست برای ذخیره مدل‌ها و نتایج
gbr_models = []
mse_gbr = []
r2_gbr = []

# آموزش و ارزیابی GBR برای هر ویژگی
for i in range(n_features):
    print(f"Training GBR for feature {i+1}")

    # ایجاد و آموزش مدل GBR
    gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)
    gbr_model.fit(gru_predictions_train, y3_train[:, i])

    # پیش‌بینی با GBR
    y_pred_gbr = gbr_model.predict(gru_predictions_test)

    # ارزیابی GBR
    mse_gbr.append(mean_squared_error(y3_test[:, i], y_pred_gbr))
    r2_gbr.append(r2_score(y3_test[:, i], y_pred_gbr))

    print(f"Mean Squared Error (Feature {i+1}): {mse_gbr[-1]}")
    print(f"R² Score (Feature {i+1}): {r2_gbr[-1]}")

# نمایش نتایج نهایی
print("\nOverall Results:")
print(f"Mean Squared Error: {mse_gbr}")
print(f"R² Score: {r2_gbr}")
```
